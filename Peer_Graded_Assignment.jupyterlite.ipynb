```python
# 1. Boxplot for MEDV
sns.boxplot(y=df['MEDV'])
plt.title('Boxplot of Median Value of Homes (MEDV)')
plt.ylabel('MEDV ($1000s)')
plt.show()
```

```python
# 2. Bar plot for CHAS
sns.countplot(x='CHAS', data=df)
plt.title('Bar Plot of Charles River Variable')
plt.xlabel('CHAS (1 = bounds river, 0 = does not)')
plt.ylabel('Count')
plt.show()

# 3. Boxplot for MEDV vs AGE_GROUP
sns.boxplot(x='AGE', y='MEDV', data=df)
plt.title('Boxplot of MEDV by AGE Group')
plt.xlabel('AGE Group')
plt.ylabel('MEDV ($1000s)')
plt.show()

# 4. Scatter Plot: NOX vs INDUS
sns.scatterplot(x='INDUS', y='NOX', data=df)
plt.title('Scatter Plot of NOX vs INDUS')
plt.xlabel('Proportion of Non-Retail Business Acres (INDUS)')
plt.ylabel('Nitric Oxide Concentration (NOX)')
plt.show()

# 5. Histogram of PTRATIO
sns.histplot(df['PTRATIO'], bins=20)
plt.title('Histogram of Pupil-Teacher Ratio (PTRATIO)')
plt.xlabel('PTRATIO')
plt.ylabel('Frequency')
plt.show()

# Statistical Tests

# T-test: CHAS and MEDV
chas_0 = df[df['CHAS'] == 0]['MEDV']
chas_1 = df[df['CHAS'] == 1]['MEDV']
t_stat, p_val_t = stats.ttest_ind(chas_0, chas_1, equal_var=False)
print("T-Test for CHAS and MEDV")
print(f"t-statistic = {t_stat:.4f}, p-value = {p_val_t:.4f}\n")

# ANOVA: AGE_GROUP and MEDV
anova = stats.f_oneway(
    df[df['AGE'] == '<=35']['MEDV'],
    df[df['AGE'] == '36-70']['MEDV'],
    df[df['AGE'] == '70+']['MEDV']
)
print("ANOVA for AGE_GROUP and MEDV")
print(f"F-statistic = {anova.statistic:.4f}, p-value = {anova.pvalue:.4e}\n")

# Pearson Correlation: NOX and INDUS
corr, p_val_corr = stats.pearsonr(df['NOX'], df['INDUS'])
print("Pearson Correlation between NOX and INDUS")
print(f"Correlation Coefficient = {corr:.4f}, p-value = {p_val_corr:.4e}\n")

# Regression: DIS and MEDV
X = sm.add_constant(df['DIS'])
model = sm.OLS(df['MEDV'], X).fit()
print(model.summary())
```
